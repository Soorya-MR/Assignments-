```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
subset_df = df[['SL', 'SW', 'PL', 'PW','Classification']]
```

```python
sns.pairplot(subset_df, hue='Classification', diag_kind='hist', height=2.5)
```

```python
plt.suptitle('Pairplot of Sepal and Petal Dimensions (SL, SW, PL, PW)', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Classification', y='SL', data=df)
plt.title('Boxplot of SL by Classification')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Classification', y='SW', data=df)
plt.title('Boxplot of SL by Classification')
plt.show()
```
